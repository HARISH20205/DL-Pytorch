```python
import torch
import torch.nn as nn
import numpy as np

class LinearRegression:
    def __init__(self):
        self.model = None
        
    def fit(self, X, y, learning_rate=0.001, epochs=100):
        X = torch.tensor(X, dtype=torch.float32).to(device='cuda')
        y = torch.tensor(y, dtype=torch.float32).view(-1, 1).to(device='cuda')
        
        self.model = nn.Linear(X.shape[1], 1).to(device='cuda')
        
        loss_fn = nn.MSELoss()
        optim = torch.optim.SGD(self.model.parameters(), lr=learning_rate)
        
        for epoch in range(epochs):
            y_pred = self.model(X)
            loss = loss_fn(y_pred, y)
            
            optim.zero_grad()
            loss.backward()
            optim.step()
            
            if (epoch + 1) % 10 == 0:
                print(f'Epoch [{epoch + 1}/{epochs}], Loss: {loss.item():.4f}')
        
    def predict(self, X):
        X = torch.tensor(X, dtype=torch.float32).to(device='cuda')
        with torch.no_grad():
            y_pred = self.model(X)
            
        return y_pred.cpu().numpy()
    
if __name__ == "__main__":
    X = np.array([[1], [2], [3], [4]], dtype=np.float32)  
    y = np.array([1,2,3,4], dtype=np.float32)  
    # np.random.seed(0)
    # X = np.random.rand(100, 1) * 10  
    # y = 3.5 * X.squeeze() + np.random.randn(100) * 0.5 

    regressor = LinearRegression()
    regressor.fit(X, y, epochs=10000, learning_rate=0.01)

    # Make predictions
    X_test = np.array([[100], [103]], dtype=np.float32)
    predictions = regressor.predict(X_test)
    print("Predictions:", predictions)  
```

```python
# Make predictions
X_test = np.array([[100], [103]], dtype=np.float32)
predictions = regressor.predict(X_test) 
print("Predictions:", predictions)  
```
